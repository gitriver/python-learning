```python
review_points_mean = reviews.points.mean()
df=reviews.points.map(lambda p: p - review_points_mean)
df.head(2)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

df2=reviews.apply(remean_points, axis='columns')
df2.head(2)
```
